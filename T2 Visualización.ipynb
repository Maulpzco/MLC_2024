### 1-. Graficar las siguientes funciones f(x). Determina el dominio a partir de la función:

$$ f(x) = x^2 -4x + 2 $$

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace (-3, 3, 10)
y = x**2 - 4*x + 2
plt.plot(x, y, "g")
plt.grid()
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Ec 1")
```

### 2-. Graficar los siguientes grupos de funciones:

$$ y1 = 2x + 1 $$
$$ y2 = -x + 4 $$
$$ y3 = x^2 $$

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace (5, -5, 101)
y1 = 2*x + 1
y2 = -x + 4
y3 = x**2
plt.plot(x, y1, "r")
plt.plot(x, y2, "g")
plt.plot(x, y3, "b")
plt.grid()
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Grupo 1")
plt.plot(x, y1, "r", label="y1")
plt.plot(x, y2, "g", label="y2")
plt.plot(x, y3, "b", label="y3")
plt.legend(loc="best")
```

### 3-. Crea una función que cree una gráfica y dibuje 'k' líneas de f(x) = sin(αx)
### La función recibe:


#### x1: valor inicial de 'x'
#### x2: valor final de 'x' 
#### n: cantidad de puntos del vector 'x' 
#### a: valor inicial de α 
#### b: valor final de α 
#### k: cantidad de líneas

```python
import numpy as np
import matplotlib.pyplot as plt

def graficas(x1, x2, n, a, b, k):
    x = np.linspace(x1, x2, n)
    α = np.linspace(a, b, k)

    for α in α:
            y = np.sin(α * x)
            plt.plot(x, y, label = "Alpha")
        
    plt.grid()
    plt.xlabel("Eje x")
    plt.ylabel("Eje y")
    plt.title("Función f(x) = sin(αx)")
    plt.legend(loc="best")
    plt.legend()
    plt.show()  

graficas(0, 2 * np.pi, 1000, 1, 3, 3)
```
